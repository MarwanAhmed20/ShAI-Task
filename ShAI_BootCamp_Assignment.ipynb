## Data Cleaning

```python
df.info()
```

```python
df.isna().sum()
```

* **We see 'BasePay', 'Benefits', 'Year', 'Notes', 'Agency' and 'Status' have missing values**
* **'Notes' and 'Status' have no value at all so lets drop it**

```python
df.drop(['Notes','Status'],axis = 1,inplace = True)
```

```python
df['Agency'].value_counts()
```

```python
#check if there is outlier in Benefits and BasePay
fig, axes = plt.subplots(1, 2)
sns.boxplot(ax=axes[0],data = df['BasePay'])
sns.boxplot(ax=axes[1],data = df['Benefits'])
```

```python
med_base = df['BasePay'].median()
med_Benefits = df['Benefits'].median()
mode_agency = df['Agency'].mode()[0]
df['BasePay'] = df['BasePay'].fillna(med_base)
df['Benefits'] = df['Benefits'].fillna(med_Benefits)
df['Agency'] = df['Agency'].fillna(mode_agency)
df['Year'] = df['Year'].fillna(2014)
```

* fillna 'BasePay' and 'Benefits' with median becuase numeric and   have outliers
* fillna  'Agency' with mode because object
* fill Year with 2014 because it is ranked

## departments are in ( ) so we will use Regex to Extract it


```python
import re
dep = [*map(lambda jobtitle: re.findall(r'\([\w|\s]*\)', jobtitle) ,df['JobTitle'].values)]
dep[:5]
```

```python
dep = ['GENERAL DEPARTMENT' if len(d) < 1 else d[0].replace('(','').replace(')','') for d in dep]
set(dep)
```

## 'FIRE DEPT' is 'FIRE DEPARTMENT' is 'Fire Department' so lets merge

```python
dep = [*map(str.upper,dep)]
set(dep)
```

```python
dep = [*map(lambda d : d.replace('FIRE DEPT','FIRE DEPARTMENT'),dep)]
set(dep)
```

```python
df['department'] = dep
df.head()
```

```python
counts = df['department'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Proportion of Employees in deparment')
plt.show()
```

```python
department_count = df['department'].value_counts().reset_index()
department_count
```

```python
others_dep = {'POLICE DEPARTMENT': 711,
         'FIRE DEPARTMENT': 77,
         'MFCC': 90,
         'SEASONAL': 116,
         'SFERS': 64}

counts = others_dep.values()
plt.pie(counts, labels=others_dep.keys(), autopct='%1.1f%%')
plt.title('Proportion of Employees in other deparment')
plt.show()
```
